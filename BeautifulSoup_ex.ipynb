```python
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = requests.get("https://www.yelp.com/search?find_desc=&find_loc=Koreatown%2C+Los+Angeles%2C+CA%2C+United+States")
url
```

```python
soup = BeautifulSoup(url.text, 'html.parser')
print(soup)
```

```python
whole_review = soup.findAll("a", attrs={"class":"css-1m051bw"})
whole_review
```

```python
restaurant_list = list()
for line in whole_review:
  text = line.get_text() #태그 안에 들어있는 텍스트 데이터만 가져오기
  restaurant_list.append(text)

print(restaurant_list)
```

```python
multi_page_result=list()
for idx, value in enumerate(restaurant_list):
  row_data = [restaurant_list[idx]]
  multi_page_result.append(row_data)

multi_page_result = pd.DataFrame(multi_page_result, columns=['name'])
multi_page_result.to_csv('yelp.csv')
```
