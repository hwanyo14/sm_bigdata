## **빅데이터 개론 2주차 과제**

```python
import numpy as np
```

1. 정규분포를 따르는 임의의 값을 가지는 행렬 X1 = (100, 10)을 만드세요.

```python
x1 = np.random.randn(100, 10)
print(x1)
```

2. <p>행렬 X1 에서 행(20~70번째), 열(1~6) 까지만 잘라내세요.</p>

```python
sliced_x1 = x1[20:70, 1:6]

print(sliced_x1)
print(sliced_x1.size)
print(sliced_x1.shape)
```

3. 잘라낸 행렬과 내적곱을 계산 할 행렬 X2 = (?, ?)를 만드세요.
    - X2의 x값은 내적곱을 계산 할 수 있는 알맞은 크기로 랜덤한 값을 사용

```python
shape_x1 = sliced_x1.shape
x2 = np.random.randn(shape_x1[1], shape_x1[0])
print(x2)
print(x2.shape)
```

4. X1 과 X2 을 서로 내적곱 Y를 구하세요.

```python
y = np.dot(sliced_x1, x2)
print(y)
```

5. Y의 역행렬을 구하세요.

```python
inv_y = np.linalg.inv(y)
print(inv_y)
```
