```python
from selenium import webdriver
import urllib.request
from bs4 import BeautifulSoup as bs
from urllib.parse import quote_plus
from selenium.webdriver.common.keys import Keys
import time
import os
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
browser = webdriver.Chrome('chromedriver', chrome_options=chrome_options)
```

```python
searchterm = 'dog'
url = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query=' + searchterm
url
```

```python
browser.get(url)
body = browser.find_element_by_css_selector('body')
```

```python
for _ in range(20):
  body.send_keys(Keys.PAGE_DOWN)
  time.sleep(1)

imgs = browser.find_elements_by_css_selector('img._image._listImage')
```

```python
if not os.path.exists(searchterm):
  os.mkdir(searchterm)

for idx, img in enumerate(imgs):
  print(idx,img.get_attribute('src'))
  imgUrl = img.get_attribute('src')
  imgName = '/content/'+searchterm+'/'+searchterm + str(idx)+'.jpg'
  imgName
  urllib.request.urlretrieve(imgUrl, imgName)
```
